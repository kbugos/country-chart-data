```python
import numpy as np
import pandas as pd
```

```python
track_df = pd.read_csv('./track_df.csv')
```

```python
track_df.head()
```

```python
track_df.drop('Unnamed: 0', axis=1, inplace = True)
```

```python
track_df.head()
```

```python
track_df.dtypes
```

```python
track_df.is_group.fillna(value=0, inplace=True)
```

```python
track_df.gender.fillna(value='male', inplace=True)
```

```python
track_df.is_group = track_df.is_group.astype(int)
```

```python
track_df.describe(include='all')
```

```python
# Key integers map to pitches using standard Pitch Class notation. There are no missing values
 
track_df.key.value_counts()
```

```python
# Mode is a binomial variable indicating if a song is major or minor. This dataset is overwhelmingly major.

track_df['mode'].value_counts()
```

```python
# Histograms of continuous variables

print(track_df.hist(column='danceability'))
print(track_df.hist(column='energy'))
print(track_df.hist(column='loudness'))
print(track_df.hist(column='speechiness'))
print(track_df.hist(column='acousticness'))
print(track_df.hist(column='liveness'))
print(track_df.hist(column='valence'))
print(track_df.hist(column='tempo'))
```

```python
# Dummy out gender, which has 3 categories (male, female, mixed)

track_df = pd.get_dummies(track_df, columns=['gender'])
```
